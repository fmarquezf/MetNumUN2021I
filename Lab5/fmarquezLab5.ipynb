# Five good performing stocks

## PayPal Holdings Inc.

### Past six months

```python
class BootCampTask(QCAlgorithm):

    def Initialize(self):
        self.SetStartDate(2020, 10, 16)
        self.SetEndDate(2021, 4, 16)
        

        #1,2. Select IWM minute resolution data and set it to Raw normalization mode
        self.pypl = self.AddEquity("PYPL", Resolution.Minute)
        self.pypl.SetDataNormalizationMode(DataNormalizationMode.Raw)

    def OnData(self, data):

        #3. Place an order for 100 shares of IWM and print the average fill price
        if not self.Portfolio.Invested:
            self.MarketOrder("PYPL", 100)
            #4. Debug the AveragePrice of IWM
            self.Debug(str(self.Portfolio["PYPL"].AveragePrice))
```
